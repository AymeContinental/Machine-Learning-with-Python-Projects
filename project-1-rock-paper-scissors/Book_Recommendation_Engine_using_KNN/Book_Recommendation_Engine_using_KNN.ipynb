```python
# ============================================
# Celda 1: Librer√≠as base
# ============================================
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix
```

```python
# ============================================
# Celda 2: Cargar datasets
# ============================================
# Los archivos est√°n disponibles en formato CSV en GitHub
books_url = "https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/books.csv"
ratings_url = "https://raw.githubusercontent.com/zygmuntz/goodbooks-10k/master/ratings.csv"

# Cargar los datos
books = pd.read_csv(books_url)
ratings = pd.read_csv(ratings_url)

print("‚úÖ Datos cargados correctamente")
print("Books:", books.shape)
print("Ratings:", ratings.shape)
display(books.head())
display(ratings.head())
```

```python
# ============================================
# Celda 3: Preprocesamiento y filtrado (versi√≥n robusta y corregida)
# ============================================

# Guardamos una copia original por seguridad
ratings_original = ratings.copy()

# Renombrar columnas para mantener coherencia
books.rename(columns={'book_id': 'ISBN', 'title': 'Book-Title'}, inplace=True)
ratings.rename(columns={'user_id': 'User-ID', 'book_id': 'ISBN', 'rating': 'Book-Rating'}, inplace=True)

# Mostrar tama√±o inicial
print("Tama√±o inicial del dataset:")
print(f"- Usuarios: {ratings['User-ID'].nunique()}")
print(f"- Libros: {ratings['ISBN'].nunique()}")

# üîΩ Filtramos suavemente (m√≠nimo 5 calificaciones por usuario/libro)
user_counts = ratings['User-ID'].value_counts()
ratings = ratings[ratings['User-ID'].isin(user_counts[user_counts >= 5].index)]

book_counts = ratings['ISBN'].value_counts()
ratings = ratings[ratings['ISBN'].isin(book_counts[book_counts >= 5].index)]

# Mostrar resultados del filtrado
print("\nDespu√©s del filtrado:")
print(f"- Usuarios restantes: {ratings['User-ID'].nunique()}")
print(f"- Libros restantes: {ratings['ISBN'].nunique()}")

# Si sigue vac√≠o, usamos una muestra del dataset original
if ratings.empty:
    print("\n‚ö†Ô∏è El dataset sigue vac√≠o. Se usar√° una muestra m√°s peque√±a sin filtrado estricto.")
    ratings = ratings_original.sample(10000, random_state=42) if len(ratings_original) > 10000 else ratings_original.copy()

# Mostrar primeras filas
display(ratings.head())
```

```python
# ============================================
# Celda 4 (robusta): Crear matriz libro-usuario y entrenar KNN
# ============================================

from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix

# --- Helpers para detectar nombres de columnas ---
def pick_col(df, candidates):
    """Devuelve el nombre real de la columna en df que coincide con candidates."""
    for c in candidates:
        if c in df.columns:
            return c
    return None

# Verificar que 'ratings' y 'books' existen
if 'ratings' not in globals() or 'books' not in globals():
    raise RuntimeError("Los DataFrames 'ratings' y/o 'books' no est√°n definidos en el entorno. Ejecuta las celdas anteriores.")

# Detectar nombres reales de columnas en ratings y books
user_col = pick_col(ratings, ['User-ID', 'user_id', 'UserID', 'userId', 'user'])
isbn_col = pick_col(ratings, ['ISBN', 'book_id', 'bookId', 'book_id'])
rating_col = pick_col(ratings, ['Book-Rating', 'rating', 'Rating', 'book_rating'])

title_col = pick_col(books, ['Book-Title', 'title', 'book_title', 'name'])

if not user_col or not isbn_col or not rating_col or not title_col:
    print("Columnas encontradas en ratings:", ratings.columns.tolist())
    print("Columnas encontradas en books:", books.columns.tolist())
    raise RuntimeError("No se pudieron detectar las columnas necesarias autom√°ticamente. Revisa los nombres de columnas.")

print(f"Usando columnas -> ratings: user='{user_col}', isbn='{isbn_col}', rating='{rating_col}'; books: title='{title_col}'")

# Hacer copia local para no modificar originales
ratings_local = ratings[[user_col, isbn_col, rating_col]].rename(columns={user_col:'User-ID', isbn_col:'ISBN', rating_col:'Book-Rating'}).copy()
books_local = books[[isbn_col, title_col]].rename(columns={isbn_col:'ISBN', title_col:'Book-Title'}).copy()

# Si ratings est√° vac√≠o, intentar usar una muestra del original (si existe ratings_original)
if ratings_local.shape[0] == 0:
    if 'ratings_original' in globals() and ratings_original.shape[0] > 0:
        print("ratings estaba vac√≠o; usando ratings_original como fallback (muestra).")
        ratings_local = ratings_original.sample(min(20000, len(ratings_original)), random_state=42)[[user_col, isbn_col, rating_col]].rename(columns={user_col:'User-ID', isbn_col:'ISBN', rating_col:'Book-Rating'}).copy()
    else:
        raise RuntimeError("El DataFrame 'ratings' est√° vac√≠o y no existe 'ratings_original' como fallback.")

# --- Filtrado suave para asegurar densidad m√≠nima (ajustable) ---
min_ratings_user = 3   # m√≠nimo calificaciones por usuario (ajusta si quieres m√°s)
min_ratings_book = 3   # m√≠nimo calificaciones por libro

user_counts = ratings_local['User-ID'].value_counts()
keep_users = user_counts[user_counts >= min_ratings_user].index
ratings_local = ratings_local[ratings_local['User-ID'].isin(keep_users)]

book_counts = ratings_local['ISBN'].value_counts()
keep_books = book_counts[book_counts >= min_ratings_book].index
ratings_local = ratings_local[ratings_local['ISBN'].isin(keep_books)]

print(f"Despu√©s del filtrado suave: registros={len(ratings_local)}, usuarios={ratings_local['User-ID'].nunique()}, libros={ratings_local['ISBN'].nunique()}")

# Si qued√≥ muy grande, limitar a top N libros y top M usuarios para evitar matrices enormes
max_books = 2000
max_users = 5000

top_books = ratings_local['ISBN'].value_counts().nlargest(max_books).index
ratings_local = ratings_local[ratings_local['ISBN'].isin(top_books)]

top_users = ratings_local['User-ID'].value_counts().nlargest(max_users).index
ratings_local = ratings_local[ratings_local['User-ID'].isin(top_users)]

print(f"Despu√©s de limitar tama√±o: registros={len(ratings_local)}, usuarios={ratings_local['User-ID'].nunique()}, libros={ratings_local['ISBN'].nunique()}")

# Merge con t√≠tulos (join limitado para evitar duplicados m√∫ltiples)
ratings_books = ratings_local.merge(books_local, on='ISBN', how='left')
ratings_books = ratings_books.dropna(subset=['Book-Title'])  # asegurar t√≠tulos presentes

# Crear matriz libro-usuario
book_user_matrix = ratings_books.pivot_table(
    index='Book-Title',
    columns='User-ID',
    values='Book-Rating',
    aggfunc='mean'   # en caso de duplicados, usar promedio
).fillna(0)

print("Dimensiones matriz libro-usuario:", book_user_matrix.shape)

# Verificar que no est√© vac√≠a
if book_user_matrix.shape[0] == 0 or book_user_matrix.shape[1] == 0:
    raise RuntimeError("La matriz libro-usuario est√° vac√≠a despu√©s del procesamiento. Ajusta los filtros o usa otro dataset.")

# Convertir a sparse
book_user_sparse = csr_matrix(book_user_matrix.values)

# Entrenar KNN
model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_jobs=-1)
try:
    model_knn.fit(book_user_sparse)
    print("‚úÖ Modelo KNN entrenado correctamente.")
    print("Matriz sparse shape:", book_user_sparse.shape)
except Exception as e:
    print("Error al entrenar KNN:", e)
    raise
```
